```python
import pandas as pd
import numpy as np
```

```python
# SE ACCEDIO A LA BASE IMPORTANDO PAQUETES PANDAS Y NUMPY
# LUEGO PD.READ.CSV IMPRIMIR LA BASE FELICIDAD
base_felicidad=pd.read_csv('https://raw.githubusercontent.com/cienciadedatos/datos-de-miercoles/master/datos/2019/2019-08-07/felicidad.csv')
```

```python
base_felicidad
```

```python
# SE UTILIZO LA INDEXACION BOOLEANA UTILIZANDO FILTRO DEL MAX Y MIN
data_bool=(base_felicidad.anio==max(base_felicidad.anio)) | (base_felicidad.anio==min(base_felicidad.anio))
```

```python
data_observacion=base_felicidad[data_bool]
data_observacion
```

```python
# MEDIDA DE TENDENCIA CENTRAL  ESCALERA DE VIDA FELICIDAD A NIVEL MUNDIAL
data_observacion.escalera_vida.mean()
```

```python
# DESVIACION ESTANDAR A NIVEL MUNDIAL
data_observacion.escalera_vida.std()
```

```python
# CON LA INDEXACION BOOLEANA TAMBIEN SE FILTRO PARA LOS PAISES CANADA Y FRANCIA
data_paises = data_observacion[data_observacion['pais'].isin(['Canadá','Francia'])]
```

```python
data_paises
```

```python
# Crear un DataFrame 
df = pd.DataFrame({
     'paises':['Canadá','Canadá','Francia','Francia'],
     'anio':['2005','2018','2005','2018'],
     'escalera_vida':  [7.418048, 7.175497, 7.093393, 6.665904]})
df
```

```python
# MEDIDA DE TENDENCIA CENTRAL CANADA
promedio_Canadá = df[df['paises'] == 'Canadá']['escalera_vida'].mean()
promedio_Canadá
```

```python
#MEDIDIA DE TENDENCIA CENTRAL FRANCIA
promedio_Francia = df[df['paises'] == 'Francia']['escalera_vida'].mean()
promedio_Francia
```

```python
#-----------ARGUMENTO TAREA FELICIDAD MEDIDA DE TENDENCIA CENTRAL-----------------
#Canadá tiene un promedio 7.29 su población tiene alta calidad de vida al promedio global 5.65, 
#es un país más feliz tienen mejores condiciones de vida a nivel social, económico, laboral.
#Canadá también en comparación con Francia con un promedio 6.87 se distingue por ser mejor 
#por tener un fuerte mercado laboral,buen sistema educativo.

#Comparar el promedio de Francia con el promedio global, después de Canadá también se distingue por ser un
# un país feliz tiene buenas condiciones de vida, también tiene un mercado altamente desarrollado.
```

```python
# Desviación estándar  de Canadá  dispersión de los datos alrededor de la media
promedio_Canadá = df[df['paises'] == 'Canadá']['escalera_vida'].std()
promedio_Canadá
```

```python
# Desviación estándar de Francia
promedio_Francia = df[df['paises'] == 'Francia']['escalera_vida'].std()
promedio_Francia
```

```python
#--------ARGUMENTO DESVIACION ESTANDAR std------------
#La desviación estándar nos muestra el resultado Canadá 0.17, Francia 0.30, a Nivel Mundial 1.12 
# entre más dispersa esta la distribución de datos más grande es su desviación estándar, 
# centrándose en la distancia de los diferentes valores respecto al valor medio.
```
